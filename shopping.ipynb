## 2.2 User Features

An user can be represented with the averaged price and discount of the items he/she purchased.

```python
train_user_feas = train_interactions
```

```python
train_user_feas.head(2)
```

```python
train_user_feas = train_user_feas.merge(train_item_feas, on="item_ID")
```

```python
train_user_feas.head(2)
```

```python
# prompt: get each user's purchase record based on train_user_feas, and calculate the purchase frequence and the total money spent, using the discount and the unit_price to get the real amount of money spent

# Group by user_ID
user_purchase_data = train_user_feas.groupby('user_id')

# Calculate purchase frequency and total money spent for each user
user_stats = user_purchase_data.agg(
    purchase_frequency=('item_ID', 'count'),
    total_spent=('unit_price', lambda x: sum(x * (1 - train_user_feas.loc[x.index, 'discount'])))
)

user_stats
```

```python
#save the user_stats to a csv file
user_stats.to_csv("user_stats.csv")
```

```python
# prompt: based on user_stats generate user clusters/categories

from sklearn.cluster import KMeans

# Load user statistics
user_stats = pd.read_csv("user_stats.csv", index_col="user_id")

# Select features for clustering
features = ['purchase_frequency', 'total_spent']
X = user_stats[features]

# Determine the optimal number of clusters (e.g., using the Elbow method)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# prompt: generate clusters for the users, k = 3, and plot the clusters, show the statistics for each cluster

# Perform KMeans clustering with k=3
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
user_stats['cluster'] = kmeans.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='purchase_frequency', y='total_spent', hue='cluster', data=user_stats, palette='viridis')
plt.title('User Clusters')
plt.xlabel('Purchase Frequency')
plt.ylabel('Total Spent')
plt.show()

# Analyze cluster statistics
cluster_stats = user_stats.groupby('cluster').agg({
    'purchase_frequency': ['mean', 'median', 'std'],
    'total_spent': ['mean', 'median', 'std']
})

cluster_stats
```

# 8. Get predictions for all users

## 8.1 Recommend items to all the users

Put all the codes relevant to one-user recommendation into a function.

```python
def get_top_K(uid, model, test_user_feas, test_item_feas, features, K=20):
    test_feas = test_user_feas[test_user_feas["user_id"] == uid]

    test_interactions = pd.DataFrame()
    test_interactions["item_ID"] = test_item_feas["item_ID"]
    test_interactions["user_id"] = uid

    test_data = test_interactions
    test_data = test_data.merge(test_item_feas, on="item_ID")
    test_data = test_data.merge(test_user_feas, on="user_id", suffixes=("_item", "_user"))

    test_X = test_data[features]

    y_pred = model.predict_proba(test_X)[:, 1]

    test_data["score"] = y_pred

    top_K = test_data.sort_values(by=["score"], axis=0, ascending=False)["item_ID"][:K]

    return list(top_K)
```

```python
results = []
for uid in test_users["user_id"]:
    result = get_top_K(uid, lr, test_user_feas, test_item_feas, features)
    result = [uid] + result
    results.append(result)
```

## 8.2 Load the results into a dataframe
